```python
from datetime import datetime

run_id = dbutils.jobs.taskValues.get(
    taskKey="bronze_ingestion",
    key="run_id",
    debugValue="unknown"
)

spark.sql(f"""
    UPDATE news_pipeline.monitoring.pipeline_runs
    SET 
        status = 'failed',
        end_time = CAST('{datetime.now()}' AS TIMESTAMP),
        error_message = 'Pipeline failed at silver or gold stage — check Workflow run logs'
    WHERE run_id = '{run_id}'
    AND status = 'running'
""")

print(f"Logged pipeline failure for run_id: {run_id}")
```
