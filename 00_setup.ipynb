```python
%sql
-- 1. Create catalog
CREATE CATALOG IF NOT EXISTS news_pipeline
MANAGED LOCATION 'abfss://news-intelligence-container@newsintelligencestgac.dfs.core.windows.net/';

-- 2. Create schemas
CREATE SCHEMA IF NOT EXISTS news_pipeline.bronze;
CREATE SCHEMA IF NOT EXISTS news_pipeline.silver;
CREATE SCHEMA IF NOT EXISTS news_pipeline.gold;

-- 3. Create volume
CREATE EXTERNAL VOLUME IF NOT EXISTS news_pipeline.bronze.raw_json_files
LOCATION 'abfss://news-intelligence-container@newsintelligencestgac.dfs.core.windows.net/bronze/raw/';

-- 4. Verify
SHOW SCHEMAS IN news_pipeline;
SHOW VOLUMES IN news_pipeline.bronze;
```

```python
%sql
-- Monitoring schema and table
CREATE SCHEMA IF NOT EXISTS news_pipeline.monitoring;

CREATE TABLE IF NOT EXISTS news_pipeline.monitoring.pipeline_runs (
    run_id STRING,
    pipeline_name STRING,
    topic STRING,
    start_time TIMESTAMP,
    end_time TIMESTAMP,
    duration_spent DOUBLE,
    status STRING,
    bronze_records INT,
    silver_records INT,
    gold_records INT,
    error_message STRING
)
```

```python
%sql
CREATE TABLE IF NOT EXISTS news_pipeline.monitoring.data_quality_metrics (
    run_id STRING,
    layer STRING,
    table_name STRING,
    rule_name STRING,
    failed_records INT,
    total_records INT,
    failure_percentage DOUBLE,
    logged_at TIMESTAMP
)
```
